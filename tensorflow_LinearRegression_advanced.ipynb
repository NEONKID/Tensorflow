# Linear Regression Advanced

좀 더 고급적인 방법으로 접근해보자

```python
# -*- coding: utf-8 -*-

import tensorflow as tf
import numpy as np

x_data = [1,2,3]
y_data = [3,5,7]

# h(x) = Wx + b
W = tf.Variable(tf.random_uniform([1], -1,1))
b = tf.Variable(tf.random_uniform([1], -1,1))
h = W * x_data + b

diff = h - y_data
loss = tf.reduce_mean(tf.square(diff)) # 차이폭을 양수화,,

alpha = tf.Variable(0.1) # Learing rate, 학습하는 시간..

optimizer = tf.train.GradientDescentOptimizer(alpha)
train = optimizer.minimize(loss) # 최대한 차이없게,,

init = tf.initialize_all_variables() # 변수 초기화,, 중요함!!

with tf.Session() as sess:
    sess.run(init)
    print ("[i] [loss] [W] [b]")
    
    for i in range(1000):
        sess.run(train)
        if i % 20 == 0:
            print (i, sess.run(loss), sess.run(W), sess.run(b))
```
