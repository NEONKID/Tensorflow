Linear Regression 학습은 numpy와 그 수학적인 개념을 Python 코드에 어떻게 적용시키는지에 대한 충분한
이해와 계산 결과에 대한 정확한 구현이 가능해야만 넘어갈 수 있는 어려운 종목이었다.

이번에는 수치상으로 보이기 어려움을 극복하기 위해 그래프를 그리는 실습을 진행한다.

```python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt

# X와 Y가 모두 1,2,3,4 (일차함수라 하나 이런걸 ;; 잘 모르겠음 수알못 ㅠㅠ)
# tensorflow에서는 LinearRegression ! (차라리 수학보다 낫다 머신 러닝 ㅋ)
plt.plot([1,2,3,4])
plt.ylabel('something')
plt.xlabel('x things')
plt.show()
```
